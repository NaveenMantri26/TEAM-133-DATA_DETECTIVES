```python
import os
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.embeddings import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_community.llms import OpenAI
from langchain.chains import RetrievalQA
```

```python
os.environ["OPENAI_API_KEY"] = "YOUR_API_KEY"
```

```python
with open("../data/learning_material.txt", "r", encoding="utf-8") as f:
    text_data = f.read()

print("Loaded learning material preview:\n")
print(text_data[:500])
```

```python
text_splitter = CharacterTextSplitter(
    separator="\n",
    chunk_size=300,
    chunk_overlap=50
)

chunks = text_splitter.split_text(text_data)
len(chunks)
```

```python
embeddings = OpenAIEmbeddings()

vectordb = Chroma.from_texts(
    texts=chunks,
    embedding=embeddings
)

retriever = vectordb.as_retriever()
```

```python
llm = OpenAI(temperature=0)

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever
)
```

```python
query = "Explain photosynthesis in simple words."
response = qa_chain.run(query)

print("Question:", query)
print("\nAI Answer:", response)
```

```python
def adaptive_question(student_correct):
    if student_correct:
        prompt = "Give me a harder question on this topic."
    else:
        prompt = "Give me an easier question with explanation."

    return qa_chain.run(prompt)

print("If student got it wrong:\n")
print(adaptive_question(student_correct=False))

print("\nIf student got it right:\n")
print(adaptive_question(student_correct=True))

```

```python
def learning_assistant():
    print("Welcome to the AI Learning Mentor! Type 'exit' to stop.")

    while True:
        user_input = input("\nAsk a question: ")
        if user_input.lower() == "exit":
            print("Goodbye!")
            break

        answer = qa_chain.run(user_input)
        print("\nAI Mentor:", answer)

learning_assistant()
```
